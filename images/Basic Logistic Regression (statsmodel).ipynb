```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv(r'C:\Users\Dylan\Desktop\ML\AddmitanceRecords.csv')
data
```

```python
data['Admitted'] = data['Admitted'].map({'Yes' : 1, 'No' : 0})
data
```

```python
y = data['Admitted']
x = data['SAT']

plt.scatter(x, y)

plt.xlabel('SAT')
plt.ylabel('Admission')
```

```python
x = sm.add_constant(x)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
```

```python
results_log.summary()
```
