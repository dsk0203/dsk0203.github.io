```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
sns.set()
```

```python
data = pd.read_csv(r'C:\Users\Dylan\Desktop\ML\SATversusGPAwithATTENDENCE.csv')
```

```python
print(data.head())
```

```python
# student attended 75% of lessons denoted with 1, if not denoted with 0.
```

```python
data['Numerical_Att'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data.head()
```

```python
data.drop(['Attendance'], axis = 1)
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Numerical_Att']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

y^ = .6439 + .0014x1 + .226x2

```python
test = pd.DataFrame({'const':1,'SAT':[1760, 1670],'Numerical_Att':[0,1]})
test = test[['const', 'SAT', 'Numerical_Att']]
```

```python
test.head()
```

```python
# feed to model
```

```python
predictions = results.predict(test)
print(predictions)
```
